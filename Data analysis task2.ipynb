```python
import pandas as pd
df = pd.read_excel(r"C:\Users\LENOVO\Desktop\Data Cleaning and Preprocessing.xlsx")
df.drop(['Observation'],axis=1,inplace=True)
df.columns
```

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
df = df[~((df< (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(df)
```

```python
print("Type of Data:",type(df))
print(df.info())
```

```python
mean_values = df.mean()
median_values = df.median()
mode_values = df.mode().iloc[0] 
std_dev_values = df.std()

# Output 
print("Mean:\n", mean_values)
print("\nMedian:\n", median_values)
print("\nMode:\n", mode_values)
print("\nStandard Deviation:\n", std_dev_values)
```

```python
outliers = df[(df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))]

# Output the outliers
print("Outliers found:")
print(outliers.dropna())
```
