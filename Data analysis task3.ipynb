```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#loading dataset
file_path= r"C:\Users\LENOVO\Desktop\householddatasettask3..xlsx"
df=pd.read_excel(file_path)
#Data inspection 
print("\n Head details")
print(df.head())
print("\n Column names")
print(df.columns)
```

```python
df.head(10)
```

```python

# Scatter plot with year against own
plt.scatter(df['year'], df['own'],color='red')

# Adding title to the plot
plt.title("Scatter Plot")

# Setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

# Displaying result
plt.show()
```

```python
# Scatter plot with year against own
plt.plot(df['year'],color="violet")
plt.plot(df['own'],color="purple")

# Adding title to the plot
plt.title("Line Chart")

# Setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

# displaying result
plt.show()
```

```python

# Scatter plot with year against own
plt.bar(df['year'], df['own'],color='orange')

# Adding title to the plot
plt.title("Scatter Plot")

# Setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

# Displaying result
plt.show()
```

```python
#histogram
plt.hist(df['income'],color="hotpink")
plt.title("Histogram")
plt.show()
```

```python
#histogram
plt.hist(df['expenditure'],color="purple")
plt.title("Histogram")
plt.show()
```

```python
#Income vs Expenditure

plt.plot(df['income'],color="blue")
plt.plot(df['expenditure'],color="red")

# Adding title to the plot
plt.title("Income Vs Expenditure")
plt.grid(True)

# Setting the x and y labels
plt.xlabel('income')
plt.ylabel('expenditure')

# displaying result
plt.show()
```
