```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Function to detect outliers using the IQR method
def detect_outliers(data_column):
    Q1 = np.percentile(data_column, 25)
    Q3 = np.percentile(data_column, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data_column[(data_column < lower_bound) | (data_column > upper_bound)]
    return outliers

# Detecting outliers
outliers_views = detect_outliers(df['views'])
outliers_likes = detect_outliers(df['likes'])
outliers_dislikes = detect_outliers(df['dislikes'])
outliers_comments = detect_outliers(df['comment_count'])

# Visualizing outliers using box plots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Views outliers
sns.boxplot(x=df['views'], ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Outliers in Views')

# Likes outliers
sns.boxplot(x=df['likes'], ax=axes[0, 1], color='green')
axes[0, 1].set_title('Outliers in Likes')

# Dislikes outliers
sns.boxplot(x=df['dislikes'], ax=axes[1, 0], color='red')
axes[1, 0].set_title('Outliers in Dislikes')

# Comment count outliers
sns.boxplot(x=df['comment_count'], ax=axes[1, 1], color='purple')
axes[1, 1].set_title('Outliers in Comment Count')

# Adjust layout
plt.tight_layout()
plt.show()

# Displaying outlier counts
len
```

```python
import pandas as pd

# Load the dataset
file_path = r"C:\Users\LENOVO\Desktop\USvideos.csv"
df = pd.read_csv(file_path)

# Compute the correlation matrix using only numerical columns
correlation_matrix = df.select_dtypes(include=['number']).corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
