```python
#1 What are the top 5 most frequent genres?
# Count the frequency of each genre
import pandas as pd
#loading dataset
file_path= r"C:\Users\LENOVO\Desktop\disney_plus_titles.csv"
df=pd.read_csv(file_path)
top_genres = df['listed_in'].value_counts().head(5)
print("Top 5 most frequent genres:")
print(top_genres)
```

```python
 #2.What is the distribution of ratings across all titles?
# Count the distribution of ratings
rating_distribution = df['rating'].value_counts()
print("Rating distribution:")
print(rating_distribution)
```

```python
#3.What is the most frequent release year for titles?
most_common_year = df['release_year'].mode()[0]
print(f"The most frequent release year: {most_common_year}")
```

```python
#4.What percentage of content is TV shows vs Movies?
content_type_counts = df['type'].value_counts(normalize=True) * 100
print("Percentage of content type (TV Shows vs Movies):")
print(content_type_counts)
```

```python
#5.Which production countries have the most titles on Disney Plus?
# Assuming the column for country is 'country'
top_countries = df['country'].value_counts().head(10)
print("Top 10 countries with the most titles:")
print(top_countries)
```

```python
#6.What is the average runtime for movies vs TV shows?
# Separate movies and TV shows
movies = df[df['type'] == 'Movie']
tv_shows = df[df['type'] == 'TV Show']

# For movies, extract duration in minutes
movies['duration_minutes'] = movies['duration'].str.extract('(\d+)').astype(float)

# For TV shows, extract the number of seasons (in case it's available as "X Seasons")
tv_shows['seasons'] = tv_shows['duration'].str.extract('(\d+)').astype(float)

# Calculate the average runtime for Movies
average_movie_runtime = movies['duration_minutes'].mean()

# Calculate the average number of seasons for TV Shows
average_tv_show_seasons = tv_shows['seasons'].mean()

print("Average runtime for Movies (in minutes):", average_movie_runtime)
print("Average number of seasons for TV Shows:", average_tv_show_seasons)
```

```python
#7 .Which directors have produced the most content on Disney Plus?
#This analysis finds the most prolific directors on the platform.
top_directors = df['director'].value_counts().head(10)
print("Top 10 directors with the most titles:")
print(top_directors)
```

```python
#8.Find the top most cast members
# Split the 'cast' column by commas and stack the resulting series
cast_series = df['cast'].dropna().str.split(', ').explode()

# Count the frequency of each cast member
top_10_cast = cast_series.value_counts().head(10)

# Display the top 10 most frequent cast members
print("Top 10 Most Frequent Cast Members:")
print(top_10_cast)
```

```python
#9. Find the number of movies released for each year in the dataset?
# Filter the dataset to include only movies
movies = df[df['type'] == 'Movie']

# Group by 'release_year' and count the number of movies for each year
movies_by_year = movies.groupby('release_year').size()

# Sort the result by year (optional)
movies_by_year = movies_by_year.sort_index()

# Display the result
print("Number of movies released each year:")
print(movies_by_year)
```

```python
#10.Find the number of movies released by top 10 countries 
# Filter the dataset to include only movies
movies = df[df['type'] == 'Movie']

# Group by 'country' and count the number of movies for each country
movies_by_country = movies.groupby('country').size()

# Sort the result in descending order to get the top 10 countries
top_10_countries = movies_by_country.sort_values(ascending=False).head(10)

# Display the top 10 countries with the number of movies released
print("Top 10 countries by the number of movies released:")
print(top_10_countries)
```

```python
#distribution of Movies vs TV Shows
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of Movies vs TV Shows
plt.figure(figsize=(7,5))
sns.countplot(data=df, x='type', palette='viridis')
plt.title('Distribution of Movies vs TV Shows')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(data=df, x='type', hue='listed_in', palette='viridis')
plt.title('Distribution of Movies vs TV Shows')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of release year:Plot the distribution of release year
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='release_year', bins=30, kde=True, color='hotpink')
plt.title('Distribution of Release Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

```python
#Distribution of rating
# Plot the distribution of content rating
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index, palette='viridis')
plt.title('Distribution of Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Extract genres for analysis
genres = df['listed_in'].str.split(', ', expand=True).stack()

# Plot the top 10 most common genres
plt.figure(figsize=(12,6))
genres.value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Common Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter the dataset to include only movies
disney_data_movies = df[df['type'] == 'Movie']

# Extract the duration in minutes by using regular expressions to find digits
disney_data_movies['duration_minutes'] = disney_data_movies['duration'].str.extract('(\d+)').astype(float)

# Plot the distribution of movie durations
plt.figure(figsize=(10,6))
sns.histplot(data=disney_data_movies, x='duration_minutes', bins=30, kde=True, color='purple')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Count')
plt.show()
```
